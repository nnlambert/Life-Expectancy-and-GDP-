```python
import pandas as pd
import numpy as np
import seaborn as sns

# We will load in the data to determine what questions we can ask
df = pd.read_csv('all_data.csv')
df.head()

print(df.describe())
```

```python
# Visualizing change in Life Expectancy over time in each country
from matplotlib import pyplot as plt
sns.set_theme(style="ticks")

sns.pairplot(df, hue="Country")
```

### Observations
